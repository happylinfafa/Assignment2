```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

diabetes = pd.read_csv('../data/diabetes.csv')
```

```
# Sampling
np.random.seed(42)
sample = diabetes.sample(n=25)

sample[['Glucose']].describe()
```

```
# Population statistics

diabetes[['Glucose']].describe()
```

```
# 98th percentile BMI

sample_p98 = np.percentile(sample['BMI'],98)
pop_p98 = np.percentile(diabetes['BMI'],98)

sample_p98, pop_p98
```

```
# Bootstrap
boot_means = []
boot_std = []
boot_p98 = []

for _ in range(500):
    s = diabetes.sample(n=150, replace=True)
    boot_means.append(s['BloodPressure'].mean())
    boot_std.append(s['BloodPressure'].std())
    boot_p98.append(np.percentile(s['BloodPressure'],98))
```

```
# Visualization
plt.hist(boot_means, bins=30)
plt.axvline(diabetes['BloodPressure'].mean(), color='red')
plt.show()
```
